# Мотивация

Хочу отдельно проговорить свою мотивацию по этой позиции и по тестовому заданию.

Задача — интеллектуальная обработка документов для проверки целевого использования льготных сельхоз-кредитов — мне интересна не абстрактно: это прикладная задача с понятной пользой (сокращение ручной проверки, снижение риска нецелевого использования субсидий), и она хорошо ложится на тот стек, с которым я хочу дальше расти. Именно поэтому, несмотря на ограниченное время, я довёл решение до рабочего состояния, покрыл тестами и явно задокументировал компромиссы (`README.md`, раздел «Компромиссы»; `RESULTS.md`), а не оставил черновик.

При этом я **очень заинтересован(-а)** в этой позиции и готов(-а) начинать уже сейчас в рамках доступных 20 часов в неделю. После завершения практики (конец августа) я перехожу на полную занятость и обязательно **доработаю** всё, что не успею закрыть за это время из-за сокращённого графика — это не будет потеряно или отложено «на потом».
